```python
# This code returns the correlation between the sp500 and nasdaq index.
%matplotlib inline
```

```python
import pandas as pd
```

```python
def get_returns(file):
    return pd.read_csv(file + '.csv', index_col=0, parse_dates=True).pct_change()



df = get_returns('../data/chapter6/sp500')

df['NDX'] = get_returns('../data/chapter6/ndx')
df['SP500'].rolling(50).corr(df['NDX'])[-200:].plot() # discards all data except last 200 rows
```
